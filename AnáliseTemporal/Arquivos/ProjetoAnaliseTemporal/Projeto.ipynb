### <font color='blue'>EVOLUÇÃO TEMPORAL DO USO DA INTERNET NO BRASIL</font>

```python
%%sql

SELECT "2000", "2001", "2002", "2003", "2004", "2005", "2006", "2007", "2008", "2009", "2010", "2011", "2012", "2013", "2014", "2015", "2016", "2017", "2018", "2019", "2020", "2021", "2022", "2023"
FROM internet_usage
WHERE "Country Name" = 'Brazil';
```

### <font color='blue'>CRESCIMENTO ABSOLUTO</font>

```python
%%sql

SELECT "Country Name", "2023" - "2000" AS crescimento_absoluto
FROM internet_usage
WHERE "Country Name" = 'Brazil';
```

### <font color='blue'>CRESCIMENTO ATUAL</font>

```python
%%sql

SELECT "Country Name",
"2023" - "2022" AS crescimento_2023
FROM internet_usage
WHERE "Country Name" = 'Brazil';
```

### <font color='blue'>MÉDIA DE CRESCIMENTO ANUAL</font>

```python
%%sql

SELECT "Country Name", (("2001" - "2000") + ("2002" - "2001") + ("2003" - "2002") + ("2004" -"2003") + ("2005" - "2004") + ("2006" - "2005") + ("2007" - "2006") + ("2008" - "2007") + ("2009" - "2008") + ("2010" - "2009") + ("2011" - "2010") + ("2012" - "2011") + ("2013" - "2012") + ("2014" - "2013") + ("2015" - "2014") + ("2016" - "2015") + ("2017" - "2016") + ("2018" - "2017") + ("2019" - "2020") + ("2020" - "2019") + ("2021" - "2020") + ("2022" - "2021") + ("2023" - "2022"))/24
AS media_crescimento_anual FROM internet_usage WHERE "Country Name" = 'Brazil';
```

### <font color='blue'>PICO DE CRESCIMENTO ANUAL</font>

```python
%%sql

SELECT "Country Name", 
MAX(MAX(("2001" - "2000")), MAX(("2002" - "2001")), MAX(("2003" - "2002")), MAX(("2004" - "2003")), MAX(("2005" - "2004")), MAX(("2006" - "2005")), MAX(("2007" - "2006")), MAX(("2008" - "2007")), MAX(("2009" - "2008")), MAX(("2010" - "2009")), MAX(("2011" - "2010")), MAX(("2012" - "2011")), MAX(("2013" - "2012")), MAX(("2014" - "2013")), MAX(("2015" - "2014")), MAX(("2016" - "2015")), MAX(("2017" - "2016")), MAX(("2018" - "2017")), MAX(("2019" - "2018")), MAX(("2020" - "2019")), MAX(("2021" - "2020")), MAX(("2022" - "2021")), MAX(("2023" - "202")))
AS pico_crescimento_anual
FROM internet_usage
WHERE "Country Name" = 'Brazil'
```

### <font color='blue'>RANKING DO ÚLTIMO ANO</font>

```python
%%sql

SELECT "Country Name", "2023" AS internet_usage_2023
FROM internet_usage

WHERE "Country Name" = 'Brazil'
ORDER BY
"2023" DESC;
```

### <font color='blue'>RANKING DO PRIMEIRO ANO</font>

```python
%%sql

SELECT "Country Name", "2000" AS internet_usage_2000
FROM internet_usage
WHERE "Country Name" = 'Brazil'
ORDER BY
"2000" DESC;
```

### <font color='blue'>MÉDIA ARITMÉTICA</font>

```python
%%sql

SELECT "Country Name",
("2000" + "2001" + "2002" + "2003" + "2004" + "2005" + "2006" + "2007" + "2008" + "2009" + "2010" + "2011" + "2012" + "2013" + "2014" + "2015" + "2016" + "2017" + "2018" + "2019" + "2020" + "2021" + "2022" + "2023")
/23.0 AS media


FROM internet_usage
WHERE "Country Name" = 'Brazil';
 
```

## <font color='blue'>CRIANDO NOVA ESTRUTURA TABULAR</font>

```python
%%sql

CREATE TABLE analise (crescimento_absoluto DECIMAL (8, 2), 
                      crescimento_atual DECIMAL (8, 2), 
                      media_de_crescimento_anual DECIMAL (8, 2), 
                      pico_de_crescimento_anual DECIMAL (8, 2),
                      ranking_do_ultimo_ano DECIMAL (8, 2),
                      ranking_do_primeiro_ano DECIMAL (8, 2),
                      media_aritmetica DECIMAL (8, 2));
                      
                    
```

```python
%%sql

INSERT INTO analise (crescimento_absoluto,
                     crescimento_atual,
                     media_de_crescimento_anual,
                     pico_de_crescimento_anual,
                     ranking_do_ultimo_ano,
                     ranking_do_primeiro_ano,
                     media_aritmetica)

SELECT "2023" - "2000", 
       "2023" - "2022",
       (("2001" - "2000") + ("2002" - "2001") + ("2003" - "2002") + ("2004" -"2003") + ("2005" - "2004") + ("2006" - "2005") + ("2007" - "2006") + ("2008" - "2007") + ("2009" - "2008") + ("2010" - "2009") + ("2011" - "2010") + ("2012" - "2011") + ("2013" - "2012") + ("2014" - "2013") + ("2015" - "2014") + ("2016" - "2015") + ("2017" - "2016") + ("2018" - "2017") + ("2019" - "2020") + ("2020" - "2019") + ("2021" - "2020") + ("2022" - "2021") + ("2023" - "2022"))/23,
       MAX(MAX(("2001" - "2000")), MAX(("2002" - "2001")), MAX(("2003" - "2002")), MAX(("2004" - "2003")), MAX(("2005" - "2004")), MAX(("2006" - "2005")), MAX(("2007" - "2006")), MAX(("2008" - "2007")), MAX(("2009" - "2008")), MAX(("2010" - "2009")), MAX(("2011" - "2010")), MAX(("2012" - "2011")), MAX(("2013" - "2012")), MAX(("2014" - "2013")), MAX(("2015" - "2014")), MAX(("2016" - "2015")), MAX(("2017" - "2016")), MAX(("2018" - "2017")), MAX(("2019" - "2018")), MAX(("2020" - "2019")), MAX(("2021" - "2020")), MAX(("2022" - "2021")), MAX(("2023" - "202"))),
       "2023",
       "2000",
       ("2000" + "2001" + "2002" + "2003" + "2004" + "2005" + "2006" + "2007" + "2008" + "2009" + "2010" + "2011" + "2012" + "2013" + "2014" + "2015" + "2016" + "2017" + "2018" + "2019" + "2020" + "2021" + "2022" + "2023")/23.0
       

FROM internet_usage
WHERE "Country Name" = 'Brazil'
ORDER BY
"2023" DESC,
"2000" DESC;
```
